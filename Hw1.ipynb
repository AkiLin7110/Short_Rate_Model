```python
import pandas_market_calendars as mcal # Define the date range
from datetime import datetime
start_date = datetime(2023, 1, 1)
end_date = datetime(2023, 4, 24)

# Create a calendar object for the NYSE exchange
nyse = mcal.get_calendar('NYSE')

# Get the schedule of trading days between start and end dates
schedule = nyse.schedule(start_date=start_date, end_date=end_date)

trading_days = schedule.index.normalize()
```

```python
import pandas as pd
import numpy as np
from scipy.optimize import fsolve
import math
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel('HW1_v2.xlsx', header = 1)
df.columns = [' ','DP/SW','Start Date','Generator','Maturity','Contract','DF','Zero Rate','Market Data','  ','TimeDelta','OD','Today_TimeDelta','Prior','DF_Cum']
df['Market Data'] =  df['Market Data']/100
```

```python
def fun(c):
    up = df['DF'][1] - df['Market Data'][c]* df['DF_Cum'][c]
    down = (1+ df['Market Data'][c] * (df['Prior'][c])/360)
    df['DF'][c] = up/down
    return  df['DF'][c]


for i in range(12,22,1):
    if i == 12:  #2Yr部分
        df['DF_Cum'][i] =  df['DF'][10] * df['TimeDelta'][10]/360 

    else:
        df['DF_Cum'][i] =  df['DF_Cum'][i-1] + df['DF'][i-1] * df['Prior'][i-1]/360  #3Y
    
    df['DF'][i] = fun(i)
    df['Zero Rate'][i] =  -np.log(df['DF'][i]) * 365/ df['Today_TimeDelta'][i]
```

```python
'''Find Zero Rate'''
def find(tmp):
    x,y = tmp[0], tmp[1]
    tmp_0 = (df['Zero Rate'][20]*df['Prior'][22] + y*df['Prior'][21])/(df['Prior'][21]+df['Prior'][22])
    tmp_1 = tmp_0 - x

    df['DF_Cum'][22] = df['DF_Cum'][21] + math.exp(-1*x*df['Prior'][21]/365) * df['Prior'][21]/360

    B = df['DF'][1] - (df['Market Data'][22]*df['DF_Cum'][22])
    C = 1+(df['Market Data'][22])*(df['Prior'][22]/360)

    tmp_2 = math.exp(-y*df['Today_TimeDelta'][22]/365)

    tmp_3 = (B/C) - tmp_2

    return tmp_1, tmp_3


x,y = fsolve(find,[0,0])
# x,y = fsolve(func,[0,0])
 
print(x)
print(y)
df['Zero Rate'][21] = x 
df['Zero Rate'][22] = y
df['DF'][21] = math.exp(-x * df['Today_TimeDelta'][21]/365)
df['DF'][22] = math.exp(-y * df['Today_TimeDelta'][22]/365)
df['DF_Cum'][22] = df['DF_Cum'][21] + df['DF'][21] * df['Prior'][21]/360
df['DF_Cum'][23] = df['DF_Cum'][22] + df['DF'][22] * df['Prior'][22]/360
```
